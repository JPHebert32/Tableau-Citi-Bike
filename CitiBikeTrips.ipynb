```python
import numpy as np
import pandas as pd
import datetime as dt
import os
```

Create YOY June 2020 DataFrame (Citi Bike Trips)

```python
yoy_june20_df = pd.DataFrame(columns = ['trip duration','start time','stop time','start station id','start station name',
                                         'start station latitude','start station longitude','end station id','end station name',
                                         'end station latitude', 'end station longitude', 'bike id','user type','birth year',
                                         'gender'])
```

```python
for fileName in os.listdir("Resources/yoy_june20"):
    if fileName.endswith(".csv"):
        temp_df=pd.read_csv("Resources/yoy_june20/"+ fileName, sep=",")
        temp_df.columns = ['trip duration','start time','stop time','start station id','start station name',
                                         'start station latitude','start station longitude','end station id','end station name',
                                         'end station latitude', 'end station longitude', 'bike id','user type','birth year',
                                         'gender']
        
        yoy_june20_df = yoy_june20_df.append(temp_df)
```

```python
yoy_june20_df["gender"].replace({0: "Unknown", 1: "Male", 2: "Female"}, inplace=True)
yoy_june20_df.head()
```

```python
yoy_june20_df.dropna(inplace=True)
print(yoy_june20_df.isnull().values.any())
```

```python
yoy_june20_df
```

```python
yoy_june20_df.to_csv('Resources/yoy_june20.csv', index=False)
```
